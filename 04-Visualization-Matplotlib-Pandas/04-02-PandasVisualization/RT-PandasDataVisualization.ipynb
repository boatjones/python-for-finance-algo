```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
```

```python
# time series data
pd.to_datetime(df1.index)
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# can make a histogram plot with 30 bins
df1['A'].hist(bins=30)
```

```python
# many different plot types are available - the same histogram is possible via this syntax
df1['A'].plot(kind='hist', bins=30)
```

```python
# yet another syntax possibility - this seems the preferred syntax
df1['A'].plot.hist(bins=30)
```

```python
# non time series data
df2.head()
```

```python
# an area plot with some transparency thrown in
df2.plot.area(alpha=0.5)
```

```python
# can plot a bar chart
df2.plot.bar()
```

```python
# can do a stacked bar chart
df2.plot.bar(stacked=True)
```

```python
# can do a pie chart of one of the columns
df2['a'].plot.pie()
```

```python
# the chart below really didn't like working with the index so had to make an ndx column
df1['ndx'] = df1.index
df1
```

```python
# making a line chart
df1.plot.line(x='ndx', y='B', figsize=(12,3), lw=0.5)
# df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)  # this just didn't work
```

```python
# this is a scatterplot where the color of the data is based off of column C in a cool to warm gradient
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# more tableau type tricks is that you can vary the size of the datapoints based off of data
df1.plot.scatter(x='A', y='B', s=df1['C']*20)
```

```python
# Candlesticks!!
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.head()
```

```python
# hexagonal plot
df.plot.hexbin(x='a', y='b', gridsize=20, cmap='coolwarm')
```

```python
# to plot the density of a column in a table
df2['a'].plot.density()
```

```python
# to plot the density of all the columns in a table
# also known as kernel density estimation: can also use "kde" instead of "density"
df2.plot.density()
```
