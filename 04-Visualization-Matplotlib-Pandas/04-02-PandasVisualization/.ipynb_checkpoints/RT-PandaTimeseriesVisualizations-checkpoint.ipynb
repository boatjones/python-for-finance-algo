```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mcdon = pd.read_csv('mcdonalds.csv', index_col='Date', parse_dates=True)
mcdon.head()
```

```python
# the issue with this chart is that the two y columns are on very different scales
mcdon.plot()
```

```python
# can chart just one of the values to avoid this
mcdon['Adj. Close'].plot()
```

```python
# same for an adjusted scale for volume
mcdon['Adj. Volume'].plot()
```

```python
# can stretch this out for a better view
mcdon['Adj. Volume'].plot()
# as a quirk, the following gives an error but afterwards, the above statment renders the chart as if 
# the below statement was being used
# mcdon['Adj. Volume'].plot(figsize(12,4))
```

```python
# can zoom in on a chart section by limiting the x and y axes and changing the linestyle and line color
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'], ylim=(20,50), ls='--', c='red')
```

```python
# you can use matplotlib's capabilities to do further work such as formatting the dates along the x axis
import matplotlib.dates as dates
```

```python
# make an index variable for a slice of the data
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx,stock)
```

```python
fig,ax = plt.subplots()

# to add a gridline for the yaxis
ax.yaxis.grid(True)

# this will auto format the dates in the x axis
fig.autofmt_xdate()

# make it a line chart
ax.plot_date(idx,stock, '-')
```

```python
fig,ax = plt.subplots()

# to add a gridline for the yaxis
ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Setting the x axes with specific major and minor locators is rarely done but detailed here for info
# locate major axes in x axis
ax.xaxis.set_major_locator(dates.MonthLocator())
# then format their display for "Mon-YYYY"  you can space them down with \n notations
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

# set minor axis in x axis on Mondays
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
# then format the display
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

# this will auto format the dates in the x axis
fig.autofmt_xdate()

# make it a line chart
ax.plot_date(idx,stock, '-')
```
