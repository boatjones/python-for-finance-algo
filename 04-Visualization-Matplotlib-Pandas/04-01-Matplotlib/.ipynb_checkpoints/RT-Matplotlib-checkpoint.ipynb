```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL APPROACH
plt.plot(x,y, 'r')  # simple plot with putting in line color specification
plt.xlabel('X Label Name')
plt.ylabel('Y Label Name')
plt.title('Title Name')
```

```python
# can have more than one plot using subplot
plt.subplot(1,2,1)
plt.plot(x,y,'r')

# the third parameter specifies the second plot
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# can do it vertically
# can have more than one plot using subplot
plt.subplot(2,1,1)
plt.plot(x,y,'r')

# the third parameter specifies the second plot
plt.subplot(2,1,2)
plt.plot(y,x,'b')
```

```python
# can make four of them
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(y,x,'b')

plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,4)
plt.plot(y,x,'y')
```

```python
# SWITCHING TO OBJECT ORIENTED METHODS
# this creates a figure object as a blank canvass
fig = plt.figure()
```

```python
# add axes to the axes via passing in a list with four values
# 1) left, 2) bottom, 3) height, 4) width as percent of canvass to take
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
```

```python
# this creates in an object oriented manner the same plot done from a functional approach
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y,'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# with this approach, there is more granular capabilities for charting & graphing
# this is the preferred approach
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'g')
axes2.set_title('Smaller Plot')
```

```python
# can make a 3 x 3 grouping of charts
fig, axes = plt.subplots(nrows=3,ncols=3)

# this automatically lays them out so that they don't overlap
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y,'r')
axes[0].set_title('Title 1')

axes[1].plot(y,x,'g')
axes[1].set_title('Title 2')

# this can be used for tightening up plot statements
plt.tight_layout()
```

```python
# fig is the object of the display
fig
```

```python
# axes is the list array of the two different chart objects
# this is why you refer to each chart via axes[0] and axes[1] above
axes
```

# Figure size and DPI

```python
# figsize is in inches 1) width, 2) height
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
# can now export a figure as any filetype you want
# specify the filename and give it an extension - it's smart enough to use that to generate the appropriate file
# supported formats: eps, jpeg, jpg, pdf, pgf, png, ps, raw, rgba, svg, svgz, tif, tiff)
fig.savefig('my_figure.jpg')

# can put in dpi resolution after filename parameter above if desired
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='x squared')
ax.plot(x,x**3, label='x cubed')

ax.set_title('My Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')

# this invokes the label function
# label parameter giving the names must be present or an error will occur 
# there are various location codes that can be put in legend() to spell out where to put the legend box
# this will default to best location (which is loc=0)
# ax.legend()
# this will put it in the center
ax.legend(loc=10)

# if no code is suitable, a tuple code for the bottom left corner can be put in like legend(loc=(0.1,0.1))
```

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# in addition to single letter color codes, can also put in full names of colors "orange" or RGB hex codes
# linewidth is another parameter able to be added.  The default is 1.  "lw" is equal to "linewidth" as parameter
# alpha is the transparency level with default being 1
ax.plot(x,y,color='#FF8C00', linewidth=8, alpha = 0.25)  # <== rgb hex code for orange
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# line style is another parameter to alter the line. Again, "ls" will equal "linestyle" as parameter
ax.plot(x,y,color='#FF8C00', lw=2, ls='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# marker is a way to put a marker at the data points of x  "x", "o", "+" are the type of options
# markersize is how to control the size of the markers.  "ms" is equal to "markersize"
# markerfacecolor is the color of hte marker's face "mfc" is equal to "markerfacecolor"
# markeredgecolor is the color of the marker's edge "mec" is equal to "markeredgecolor"
# markeredgewidth is the width of the marker's edge "mew" is equal to "markeredgewidth"
# see Matplotlib Concepts Lecture notebook for more documentation on this
ax.plot(x,y,color='blue', marker='o', ms=15, mfc='yellow', mec='red', mew=3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,marker='o')

# can set a limit on the x-axis to focus on a certain area of the chart
ax.set_xlim([0,2])

# this leaves the full range to 25 on the y-axis so possible to limit on the y-axis also
ax.set_ylim([0,5])
```
