```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_air = pd.read_csv('airline_passengers.csv', index_col='Month')
df_air.head()
```

```python
# show index and show what type of data it is
df_air.index
```

```python
df_air.info()
```

```python
# get rid of rows missing dates
df_air.dropna(inplace=True)
```

```python
# now you can convert the index to a date
df_air.index = pd.to_datetime(df_air.index)
```

```python
# now the index shows as a datetime format
df_air.index
```

```python
df_air.head()
```

```python
# calculate a new column for a 6-month simple moving average data every month so the period is 6
df_air['6Mo-SMA'] = df_air['Thousands of Passengers'].rolling(window=6).mean()
df_air.tail()
```

```python
# get rid of duplicate column
df_air.drop('6MO-SMA', axis=1, inplace=True)
```

```python
df_air['12Mo-SMA'] = df_air['Thousands of Passengers'].rolling(window=12).mean()
```

```python
df_air.plot(figsize=(12,8))
```

```python
# now calculate a 12 month Exponentially Weighted Moving Average (EWMA)
df_air['EWMA-12'] = df_air['Thousands of Passengers'].ewm(span=12).mean()
```

```python
df_air.tail()
```

```python
# now compare the data visually
df_air[['Thousands of Passengers', '12Mo-SMA', 'EWMA-12']].plot(figsize=(12,8))
```
