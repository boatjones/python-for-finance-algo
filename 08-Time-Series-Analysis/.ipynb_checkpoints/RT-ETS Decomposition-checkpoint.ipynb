```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_air = pd.read_csv('airline_passengers.csv', index_col='Month')
```

```python
df_air.dropna(inplace=True)
```

```python
df_air.index = pd.to_datetime(df_air.index)
```

```python
df_air.plot()
```

```python
# bring in the time series analysis portion to be able to analyze seasonal data
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# the main choices are "multiplicative" and "additive" - here the data is not linear but more exponential
# indicating that multiplicative is best
result = seasonal_decompose(df_air['Thousands of Passengers'], model = 'multiplicative')
```

```python
# a plot showing just the seasonal effect on the data
result.seasonal.plot()
```

```python
# so this is the chart plot of the data with the seasonality removed
result.trend.plot()
```

```python
# this will chart all of the factors in the analysis
# however this will also yield a duplicate range of charts
result.plot()
```

```python
# you can avoid this by the following
fig = result.plot()
```

```python
# calculate using the additive model
result2 = seasonal_decompose(df_air['Thousands of Passengers'], model = 'additive')
```

```python
# a set of charts using the additive model
fig = result2.plot()
```
